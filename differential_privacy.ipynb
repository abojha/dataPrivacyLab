```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/DP_Lab/preprocessed_adult.csv")
```

```python
quasi_identifiers = ["age", "education", "relationship", "sex", "race"]


df["income"] = df["income"].apply(lambda x: np.random.randint(20000, 90000) if x == "<=50K" else np.random.randint(50000, 150000))
sensitive_attribute = "income"
```

```python
df
```

```python
sensitive_attribute = "income"

# Define sensitivity (max change in `income` for one record change)
sensitivity = df[sensitive_attribute].max() - df[sensitive_attribute].min()

# Apply Laplace noise
epsilon_used = 1.0  # Theoretical epsilon used for noise
df["noisy_income"] = df[sensitive_attribute] + np.random.laplace(loc=0, scale=sensitivity/epsilon_used, size=len(df))

# Estimate actual noise added
noise = df["noisy_income"] - df[sensitive_attribute]
avg_noise = np.abs(noise).mean()

# Compute estimated epsilon
estimated_epsilon = sensitivity / avg_noise

print(f"Estimated Îµ (epsilon) for Differential Privacy: {estimated_epsilon:.2f}")
```

 **If estimated Îµ is close to the intended Îµ (e.g., 1.0), it means the Laplace mechanism is correctly implemented.
If Îµ is too high (> 2.0), it means the privacy guarantee is weak (too much information leakage).
If Îµ is too low (< 0.5), the data might be too noisy, reducing utility.**
