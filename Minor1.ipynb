### K-**Anonymity**

```python
# checking whether K-anonymous or not
def is_anony(df, k):
  count = df.groupby(quasi_identifiers).size()
  count_k = count[count < k]

  return len(count_k)

```

```python
# function for applying the k-anonymization

def a_age(age):
  if age < 30:
    return "young"
  elif age < 50:
    return "middle_aged"
  else: return "old"

def a_education(education):
  if education == "Masters" or education == "Bachelors":
    return "Higher_Studies"

  else:
    return "Lower_Studies"

def a_sex(sex):
  return "Any Sex"

def a_race(race):
  return "Any Race"

def a_relationship(relationship):
  rel = ["Husband", "Wife", "Own-child"]
  if relationship in rel:
    return "My Family"
  else : return "Non-Family"
```

```python
# applying functions to the dataset
released_data['age'] = released_data['age'].apply(a_age)
released_data['relationship'] = released_data['relationship'].apply(a_relationship)
released_data['education'] = released_data['education'].apply(a_education)
released_data['sex'] = released_data['sex'].apply(a_sex)
released_data['race'] = released_data['race'].apply(a_race)
```

```python
# now check the k_anonymization again
val = is_anony(released_data, 3)
val
```

```python
if val == 0:
  print("Data Anonymized Successfully")
else:
  print("Data is not Anonymized")
```

### **L-Diversity**

```python
# Implement L-diversity now
def entropyL(df):
  groups = df.groupby(quasi_identifiers)
  l = 100
  for names, grps in groups:
    # print("new group")
    counts = grps.iloc[:,-1].value_counts()
    rows = len(grps)
    log_val = 0.0
    for item in counts:
      log_val += -(item/rows)*np.log2(item/rows)
      # print(log_val)
    l = min(l, log_val)
  return l

#Implemented the L-diversity here to check how much diverse the sensitive_attributes with the quasi_identifiers_group
```

```python
print("entropy = ",entropyL(merged_df)) # the entropy is 0.13 means in the dataset 13% variations is there with respect to  particular quasi_identifer group
```

### **T-closeness**

```python
#implement t-closeness
def earth_movers_distance(dist1, dist2):
    return np.sum(np.abs(np.cumsum(np.sort(dist1)) - np.cumsum(np.sort(dist2))))

def sensitive_attribute_distribution(class_data, sensitive_attr):
    return class_data[sensitive_attr].value_counts(normalize=True)
```

```python
def t_closeness(df, quasi_identifiers, sensitive_attr, t_threshold):
    eq_classes = df.groupby(quasi_identifiers)

    overall_dist = sensitive_attribute_distribution(df, sensitive_attr)

    for _, class_data in eq_classes:
        class_dist = sensitive_attribute_distribution(class_data, sensitive_attr)
        emd = earth_movers_distance(overall_dist, class_dist)

        if emd > t_threshold:
            print(f"Warning: t-closeness violated for equivalence class: {quasi_identifiers}")
            print(f"EMD: {emd} exceeds threshold of {t_threshold}")
        else:
            print(f"t-closeness satisfied for equivalence class: {quasi_identifiers}")
            print(f"EMD: {emd}")
```

```python
# Set the t-closeness threshold
t_threshold = 0.1

# Run t-closeness
t_closeness(merged_df, quasi_identifiers, 'income', t_threshold)
```

#**k-Anonymity**
in k-anonymity we want that atleast k-tuples should be there whose quasi-identifiers is identical, so that the corresponding sensitive attributes may be preserved to disclose if someone knows about the quasi-identifiers...
but the problem with k-anonymity is that if say there is one group of quasi_identifiers where which has 5 rows and all 5 having same sensitive value then it is easy to identify the sensitive value of tha individual if attacker knows the information of quasi-identifier....

***means the K-anonymity does not care about the variation in the sensitive attributes***


# **L-Diversity**
So we introduce the L-Diversity which check the sensititve values should have some diversity so that in a equivalence class of quasi-identifiers the corresponding sensitive attributes have some variations so that the individual information can be preserved.

**l-diversity does not consider semantic meanings of sensitive values l-diversity is insufficient to prevent attribute disclosure.**


# **t-closeness**
requires that the distribution of a sensitive attribute in any eq. class is close to the distribution of a sensitive attribute in the overall table.
means if an sensitive attributes that is more frequently occuring it should have same prob distribution in equaivalence classes also.
